Reading image 

```python
img = cv2.imread('/content/pit-bull-featured.jpg') #image is in the form of bgr, have to convert it into rgb
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) #image converted to rgb
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray) #image converted to rgb
```

```python
#Finding the shape of the image

print(img_rgb.shape)
```

```python
#reshaping image

width,height = 300 , 200
img_resize = cv2.resize(img_rgb,(width,height))
plt.imshow(img_resize)
```

```python
print(img_resize.shape)
```

```python
#Drawing border

img_border = cv2.copyMakeBorder(img_resize, 10, 10, 5, 5, cv2.BORDER_ISOLATED)
plt.imshow(img_border)
```

```python
#Rotating image

img_rotate = cv2.rotate(img_resize, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rotate)
```

```python
#Saving a image

cv2.imwrite('output.jpg', img_resize) #this will save the image 
```

```python
#Blur image

img_blur = cv2.GaussianBlur(img_rgb,(13,13),cv2.BORDER_DEFAULT) #use odd numbers for blur value
plt.imshow(img_blur)
```
